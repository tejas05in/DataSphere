# <b>`DBSCAN`</b>
## density-based spatial clustering of applications with noise

For each instance, the algorithm counts how many instances are 
located within a small distance ε (epsilon) from it. This region is calle 
the instance’s ε-neighborhoo .
If an instance has at least min_samples instances in its ε-neighborh od
(including itself), then it is considered a core instance. In other wo ds,
core instances are those that are located in dense reg ons.
All instances in the neighborhood of a core instance belong  o the
same cluster. This neighborhood may include other core ins ances;
therefore, a long sequence of neighboring core instances forms   singlecluster.
Any instance that is not a core instance and does not have  ne in its
neighborhood is considereyd an anomal

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
X , y = make_moons(n_samples=1000 , noise=0.05)
plt.scatter(X[:,0],X[:,1],c=y,marker='.')
plt.xlabel('X[0]')
plt.ylabel("X[1]")
plt.title('Actual clusters')
plt.show()
```

```python
X
```

```python
dbscan  = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
```

```python
dbscan.labels_
# -1 = outliers
```

```python
# The indices of the core instances are available in the core_sample_indices_ 
dbscan.core_sample_indices_
```

```python
# components
# copy of each core sample found during training
dbscan.components_
```

```python
labels = dbscan.labels_
plt.scatter(X[:,0],X[:,1],c=labels,marker='.')
plt.xlabel('X[0]')
plt.ylabel("X[1]")
plt.title('eps = 0.05, min_samples=5')
plt.show()
```

```python
dbscan  = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
labels = dbscan.labels_
plt.scatter(X[:,0],X[:,1],c=labels,marker='.')
plt.xlabel('X[0]')
plt.ylabel("X[1]")
plt.title('eps = 0.2, min_samples=5')
plt.show()
```

Surprisingly, the DBSCAN class does not have a predict() method, although 
it has a fit_predict() method. In other words, it cannot predict whic 
cluster a new instance belongs to. This decision was made because differe t
classification algorithms can be better for different tasks, so the authors
decided to let the user choose which one to use. Moreover, it’s not har  to
implement. For example, let’s train a KNeighborsClassifier:

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(dbscan.components_,dbscan.labels_[dbscan.core_sample_indices_])
```

given a few new instances, we can predict which clusters they most llikely belong to and even estimate a probability for each cluster

```python
X_new = np.array([[-0.5,0],[0,0.5],[1,0.1],[2,1]])
knn.predict(X_new)
```

```python
## Probabilities of prediction
knn.predict_proba(X_new)
```
